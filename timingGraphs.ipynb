# GPU

```python
# Block_size
x = [8,16,32,64,128]
# ms averaged over 5 runs
y_half = [8.186688,6.725152,5.985824,5.976416,5.96336]
y_full = [16.501633,13.661536,12.025312,11.71424,11.71552]
y_double = [31.447968,25.027489,23.437696,21.492865,21.506847]
y_quad = [58.553249,46.269791,39.933792,38.798527,39.203297]

y_half_speedup = [ (y_half[0])/x for x in y_half]
y_full_speedup = [ (y_full[0])/x for x in y_full]
y_double_speedup = [(y_double[0])/x for x in y_double]
y_quad_speedup = [(y_quad[0])/x for x in y_quad]

plt.title('GPU Speedups')
plt.xlabel('Number of Threads')
plt.ylabel('speedup')
plt.plot(x, y_half_speedup, label="half")
plt.plot(x, y_full_speedup, label="full")
plt.plot(x, y_double_speedup, label="double")
plt.plot(x, y_quad_speedup, label="quad")
plt.legend()
if SAVE_FIGURES:
  plt.savefig("GPUspeedup.png", bbox_inches="tight")
plt.show()



plt.title('GPU Half Data')
plt.xlabel('Block-Size')
plt.ylabel('Time in ms')
plt.plot(x, y_half, label="half")
plt.legend()
if SAVE_FIGURES:
  plt.savefig("GpuHalf.png", bbox_inches="tight")
plt.show()

plt.title('GPU Full Data')
plt.xlabel('Block-Size')
plt.ylabel('Time in ms')
plt.plot(x, y_full, label="full")
plt.legend()
if SAVE_FIGURES:
  plt.savefig("GpuFull.png", bbox_inches="tight")
plt.show()

plt.title('GPU Double Data')
plt.xlabel('Block-Size')
plt.ylabel('Time in ms')
plt.plot(x, y_double, label="double")
plt.legend()
if SAVE_FIGURES:
  plt.savefig("GpuDouble.png", bbox_inches="tight")
plt.show()

plt.title('GPU Quad Data')
plt.xlabel('Block-Size')
plt.ylabel('Time in ms')
plt.plot(x, y_quad, label="quad")
plt.legend()
if SAVE_FIGURES:
  plt.savefig("GpuQuad.png", bbox_inches="tight")
plt.show()
```

# MP

```python
# threads
x = [2,4,8,16]

y_half = [36,38,42,45]
y_full = [71,76,80,84]
y_double = [146,152,160,162]
y_quad = [293,306,313,319]

y_half_speedup = [1] + [ x/(x + 1) for x in y_half[1:]]
y_full_speedup = [1] + [ x/(x + 1) for x in y_full[1:]]
y_double_speedup = [1] + [x/(x + 1) for x in y_double[1:]]
y_quad_speedup = [1] + [x/(x + 1) for x in y_quad[1:]]

y_half_speedup = [ (y_half[0])/x for x in y_half]
y_full_speedup = [ (y_full[0])/x for x in y_full]
y_double_speedup = [(y_double[0])/x for x in y_double]
y_quad_speedup = [(y_quad[0])/x for x in y_quad]

plt.title('GPU Speedups')
plt.xlabel('Number of Threads')
plt.ylabel('speedup')
plt.plot(x, y_half_speedup, label="half")
plt.plot(x, y_full_speedup, label="full")
plt.plot(x, y_double_speedup, label="double")
plt.plot(x, y_quad_speedup, label="quad")
plt.legend()
plt.show()
if SAVE_FIGURES:
  plt.savefig("MPspeedup.png", bbox_inches="tight")



plt.title('MP Half Data')
plt.xlabel('Number of Threads')
plt.ylabel('Time in ms')
plt.plot(x, y_half, label="half")
plt.legend()
plt.show()
if SAVE_FIGURES:
  plt.savefig("MPHalf.png", bbox_inches="tight")

plt.title('MP Full Data')
plt.xlabel('Number of Threads')
plt.ylabel('Time in ms')
plt.plot(x, y_full, label="full")
plt.legend()
plt.show()
if SAVE_FIGURES:
  plt.savefig("MPFull.png", bbox_inches="tight")

plt.title('MP Double Data')
plt.xlabel('Number of Threads')
plt.ylabel('Time in ms')
plt.plot(x, y_double, label="double")
plt.legend()
plt.show()
if SAVE_FIGURES:
  plt.savefig("MPDouble.png", bbox_inches="tight")

plt.title('MP Quad Data')
plt.xlabel('Number of Threads')
plt.ylabel('Time in ms')
plt.plot(x, y_quad, label="quad")
plt.legend()
plt.show()
if SAVE_FIGURES:
  plt.savefig("MPQuad.png", bbox_inches="tight")
```
