```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
```

```python
serial_csv = pd.read_csv("serialResults.csv")
serial_csv
```

```python
# axes instance
fig = plt.figure(figsize=(6,6))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

# get colormap from seaborn
cmap = ListedColormap(sns.color_palette("husl", 256).as_hex())

# plot
sc = ax.scatter(serial_csv.energy, serial_csv.tempo, serial_csv.loudness, c=serial_csv.centroid, marker='o', cmap=cmap, alpha=1)
ax.set_xlabel('energy')
ax.set_ylabel('tempo')
ax.set_zlabel('loudness')

# legend
plt.legend(*sc.legend_elements(), bbox_to_anchor=(1.05, 1), loc=2)

# save
#plt.savefig("serial_results")
```
